```python
def normalize_text(text: str = "Не все те повара, что с длинными ножами ходят."):
    norm_text = text.replace(" ", "").lower().replace(",", "зпт").replace(".", "тчк")
    return norm_text

def output_text(text: str, not_print = True) -> str:
    res = ""
    for ind, sim in enumerate(text):
        if ind % 5 == 0 and ind != 0:
            res += " "
        res += sim
    if not not_print: print(res)
    return res

ALPHABET = "абвгдежзийклмнопрстуфхцчшщъыьэюя"
```

```python
def caesar_cipher(text: str, shift: int) -> str:
    res = []
    n = len(ALPHABET)
    for ch in text:
        if ch in ALPHABET:
            res.append(ALPHABET[(ALPHABET.index(ch) + shift) % n])
        else:
            res.append(ch)
    return "".join(res)


def trithemius_cipher(text: str) -> str:
    res = []
    n = len(ALPHABET)
    for i, ch in enumerate(text):
        if ch in ALPHABET:
            res.append(ALPHABET[(ALPHABET.index(ch) + i) % n])
        else:
            res.append(ch)
    return "".join(res)


def belazo_cipher(text: str, key: str) -> str:
    res = []
    n = len(ALPHABET)
    key = normalize_text(key)
    for i, ch in enumerate(text):
        if ch in ALPHABET:
            shift = ALPHABET.index(key[i % len(key)])
            res.append(ALPHABET[(ALPHABET.index(ch) + shift) % n])
        else:
            res.append(ch)
    return "".join(res)


def vigenere_autokey_plaintext(text: str, key: str) -> str:
    res = []
    n = len(ALPHABET)
    key_stream = list(normalize_text(key))

    for ch in text:
        if ch in ALPHABET:
            shift = ALPHABET.index(key_stream.pop(0))
            cipher_ch = ALPHABET[(ALPHABET.index(ch) + shift) % n]
            res.append(cipher_ch)
            key_stream.append(ch)
        else:
            res.append(ch)
    return "".join(res)


def vigenere_autokey_ciphertext(text: str, key: str) -> str:
    res = []
    n = len(ALPHABET)
    key_stream = list(normalize_text(key))

    for ch in text:
        if ch in ALPHABET:
            shift = ALPHABET.index(key_stream.pop(0))
            cipher_ch = ALPHABET[(ALPHABET.index(ch) + shift) % n]
            res.append(cipher_ch)
            key_stream.append(cipher_ch)
        else:
            res.append(ch)
    return "".join(res)


def encrypt_all(text: str) -> dict:
    text_norm = normalize_text(text)

    result = {
        "Атбаш": text_norm[::-1],
        "Цезарь классический": caesar_cipher(text_norm, 3),
        "Цезарь со смещением 5": caesar_cipher(text_norm, 5),
        "Тритемий": trithemius_cipher(text_norm),
        "Белазо": belazo_cipher(text_norm, "гермес"),
        "Виженер с самоключом": vigenere_autokey_plaintext(text_norm, "магия"),
        "Виженер с самоключом по шифротексту": vigenere_autokey_ciphertext(text_norm, "м"),
    }

    return {k: output_text(v) for k, v in result.items()}


if __name__ == "__main__":
    text = "не все те повара, что с длинными ножами ходят."
    encrypted = encrypt_all(text)

    for name, cipher in encrypted.items():
        print(f"'{name}': '{cipher}'")
```
