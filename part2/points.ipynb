```python
p = 11
a = 4
b = 3

O = None  # точка на бесконечности


def inv(x):
    return pow(x, -1, p)


def add(P, Q):
    if P is O:
        return Q
    if Q is O:
        return P

    x1, y1 = P
    x2, y2 = Q

    if x1 == x2 and (y1 + y2) % p == 0:
        return O

    if P != Q:
        lam = ((y2 - y1) * inv(x2 - x1)) % p
    else:
        lam = ((3 * x1 * x1 + a) * inv(2 * y1)) % p

    x3 = (lam * lam - x1 - x2) % p
    y3 = (lam * (x1 - x3) - y1) % p

    return (x3, y3)


# 1. Все точки кривой
points = []
for x in range(p):
    rhs = (x**3 + a*x + b) % p
    for y in range(p):
        if (y*y) % p == rhs:
            points.append((x, y))

# 2. Для каждой точки считаем кратные
columns = {}
max_len = 0

for P in points:
    col = []
    Q = P
    while Q is not O:
        col.append(Q)
        Q = add(Q, P)
    columns[P] = col
    max_len = max(max_len, len(col))

# 3. Вывод таблицы
headers = list(columns.keys())

print("Точки кривой:")
for h in headers:
    print(f"{h}", end="\t")
print()

print("-" * 80)

for i in range(max_len):
    for h in headers:
        col = columns[h]
        if i < len(col):
            print(f"{col[i]}", end="\t")
        else:
            print(" ", end="\t")
    print()
```
