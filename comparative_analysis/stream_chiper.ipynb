```python
import matplotlib.pyplot as plt
from collections import Counter

def plot_byte_frequency(hex_string, title):
    data = bytes.fromhex(hex_string)
    freq = Counter(data)

    x = range(256)
    y = [freq.get(i, 0) for i in x]

    plt.figure(figsize=(10, 3))
    plt.bar(x, y)
    plt.xlabel("Byte value")
    plt.ylabel("Frequency")
    plt.title(title)
    plt.show()

hex_rc4 = hex_str_RC4
hex_chacha20 = hex_str_ChaCha20

plot_byte_frequency(hex_rc4, "RC4 byte frequency analysis")
plot_byte_frequency(hex_chacha20, "ChaCha20 byte frequency analysis")
```

```python
import matplotlib.pyplot as plt
from collections import Counter

def plot_bigram_frequency(hex_string, title, top_n=50):
    data = bytes.fromhex(hex_string)
    bigrams = Counter(zip(data, data[1:]))

    common = bigrams.most_common(top_n)
    labels = [f"{a:02x}{b:02x}" for (a, b), _ in common]
    values = [v for _, v in common]

    plt.figure(figsize=(10, 3))
    plt.bar(range(len(values)), values)
    plt.xticks(range(len(labels)), labels, rotation=90)
    plt.xlabel("Bigram (hex)")
    plt.ylabel("Frequency")
    plt.title(title)
    plt.tight_layout()
    plt.show()

hex_rc4 = hex_str_RC4
hex_chacha20 = hex_str_ChaCha20

plot_bigram_frequency(hex_rc4, "RC4 bigram analysis")
plot_bigram_frequency(hex_chacha20, "ChaCha20 bigram analysis")
```

```python
import matplotlib.pyplot as plt
from collections import Counter

def plot_trigram_frequency(hex_string, title, top_n=40):
    data = bytes.fromhex(hex_string)
    trigrams = Counter(zip(data, data[1:], data[2:]))

    common = trigrams.most_common(top_n)
    labels = [f"{a:02x}{b:02x}{c:02x}" for (a, b, c), _ in common]
    values = [v for _, v in common]

    plt.figure(figsize=(10, 3))
    plt.bar(range(len(values)), values)
    plt.xticks(range(len(labels)), labels, rotation=90)
    plt.xlabel("Trigram (hex)")
    plt.ylabel("Frequency")
    plt.title(title)
    plt.tight_layout()
    plt.show()

hex_rc4 = hex_str_RC4
hex_chacha20 = hex_str_ChaCha20

plot_trigram_frequency(hex_rc4, "RC4 trigram analysis")
plot_trigram_frequency(hex_chacha20, "ChaCha20 trigram analysis")
```
