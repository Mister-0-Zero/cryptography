```python
import pandas as pd
```

```python
text = "Не все те повара, что с длинными ножами ходят."
```

```python
ALF = list(sorted("абвгдежзийклмнопрстуфхцчшщъыьэюя"))
```

```python
def normalize_text(text):
    text = text.lower()
    text = text.replace("ё", "е")
    text = text.replace(",", "зпт")
    text = text.replace(".", "тчк")
    text = text.replace(" ", "")
    return text
```

```python
def get_bits_for_encryption(text: str, start_val: str, skrembler: str) -> list[str]:
    print("-----------Получаем биты для шифрования----------------")
    bits_seq = []
    reg_val = []
    val = start_val

    print("Получаем значения регистра:")
    for i in range(len(text) * 5):  # 5 - т.к. в 1 символе 5 бит
        new_bit = str(sum([int(val_) for val_, mask in zip(val, skrembler) if mask == '1']) % 2)
        reg_val.append(val)
        bits_seq.append(val[-1])
        val = str(new_bit) + val[:-1]

    print("Длина уникальных значений регистра:")
    print(len(set(reg_val)))

    max_len = 2 ** len(skrembler) - 1
    print("Сколько по хорошему должно быть уникальных значений:")
    print(max_len)

    return bits_seq, reg_val
    
```

```python
def generate_table_for_task1(reg_val, bits_seq, len_t1):
    df = pd.DataFrame({"Номер сдвига": [i for i in range(1, len(reg_val[:len_t1]) + 1)], 
                       "Значение регистра": reg_val[:len_t1], "Выходной бит": bits_seq[:len_t1]})
    display(df)
```

```python
def generate_table_for_task2(reg_val1, bits_seq1, reg_val2, bits_seq2, len_t2):
    df = pd.DataFrame({"Номер сдвига": [i for i in range(1, len_t2 + 1)],
                       "Значение регистра1": reg_val1[:len_t2],
                       "Значение регистра2": reg_val2[:len_t2],
                       "Выходной бит1": bits_seq1[:len_t2],
                       "Выходной бит2": bits_seq2[:len_t2],
                       "Выходной бит": [(int(val1) + int(val2)) % 2 for val1, val2 in zip(bits_seq1[:len_t2], bits_seq2[:len_t2]) ]})
    display(df)
```

```python
def get_gamma(text, bits_seq1, bits_seq2):
    gamma = []
    for i in range(0, len(text) * 5, 5):
        gamma_ = ''
        for x in range(5):
            sl1 = int(bits_seq1[i + x])
            sl2 = int(bits_seq2[i + x])
            summ = sl1 + sl2
            gamma_ += str(summ % 2)
        gamma.append(gamma_)
    return gamma
```

```python
def encryption(text, gamma):
    encrypted_text = ''
    for char, g in zip(text, gamma):
        ind = ALF.index(char)
        plus_ind = int(g, 2)
        res_ind = (ind + plus_ind) % len(ALF)
        encrypted_text += ALF[res_ind]
    return encrypted_text
```

```python
def main(text, skrembler="101", start_val="110", len_t1=25, len_t2=35):
    text = normalize_text(text)
    print("-----------Задание 1----------------")
    bits_seq1, reg_val1 = get_bits_for_encryption(text, start_val, skrembler)
    generate_table_for_task1(reg_val=reg_val1, bits_seq=bits_seq1, len_t1=len_t1)
    print("-----------Задание 2----------------")
    bits_seq2, reg_val2 = get_bits_for_encryption(text, "1111", skrembler="1001")
    generate_table_for_task2(reg_val1=reg_val1, bits_seq1=bits_seq1, reg_val2=reg_val2, bits_seq2=bits_seq2, len_t2=len_t2)
    gamma = get_gamma(text, bits_seq1, bits_seq2)
    print("Гамма для шифрования (первые 5):")
    print(gamma[:5])
    print(text)
    encrypted_text = encryption(text, gamma)
    print("Зашифрованный текст:")
    print(encrypted_text)
```

```python
main(text, skrembler="10111", start_val="11111")
```
