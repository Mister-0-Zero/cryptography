```python
nums_d = []

for i in range(len(nums) - 3):
    num_=  [nums[i], nums[i + 1], nums[i + 2]]
    count = 0
    for j in range(len(nums) - 3):
        num__=  [nums[j], nums[j + 1], nums[j + 2]]
        if num_ == num__:
            count += 1
    if count > 1:
        # print(f'num: {num_}, count: {count}')
        if num_ not in nums_d:
            nums_d.append(num_)
```
